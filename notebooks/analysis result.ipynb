```python
import pandas as pd
```

```python
data = pd.read_csv('./Predict_result/typeIII_submission_full_collection.csv')
data 
```

```python
Epik_Scan = data[(data['name'] == 'Epik Scan') & (data['pKa ID'] != 'SM18_pKa3')&(data['pKa ID'] != 'SM14_pKa1')]
Epik_Scan = Epik_Scan.sort_values(by=['Molecule ID','pKa (exp)'])
print(len(Epik_Scan))
Epik_Scan
```

```python
ACD = data[(data['name'] == 'ACD/pKa Classic') & (data['pKa ID'] != 'SM18_pKa3')&(data['pKa ID'] != 'SM14_pKa1')]
ACD = ACD.sort_values(by=['Molecule ID','pKa (exp)'])
print(len(ACD))
ACD
```

```python
MoKa = data[(data['name'] == 'MoKa') & (data['pKa ID'] != 'SM18_pKa3')&(data['pKa ID'] != 'SM14_pKa1')]
MoKa = MoKa.sort_values(by=['Molecule ID','pKa (exp)'])
print(len(MoKa))
MoKa
```

```python
Epik_Micro = data[(data['name'] == 'Epik Microscopic') & (data['pKa ID'] != 'SM18_pKa3')&(data['pKa ID'] != 'SM14_pKa1')]
Epik_Micro = Epik_Micro.sort_values(by=['Molecule ID','pKa (exp)'])
print(len(Epik_Micro))
Epik_Micro
```

```python
Chemicalize = data[(data['name'] == 'Chemicalize v18.23 (ChemAxon MarvinSketch v18.23)') & (data['pKa ID'] != 'SM18_pKa3')&(data['pKa ID'] != 'SM14_pKa1')]
Chemicalize= Chemicalize.sort_values(by=['Molecule ID','pKa (exp)'])
print(len(Chemicalize))
Chemicalize
```

```python
pka_graph = pd.read_csv('./Predict_result/SAMPL6_result.csv')
pka_graph= pka_graph.sort_values(by=['Molecule ID','pKa (exp)'])
print(len(pka_graph))
order = ['smiles', 'Molecule ID', 'pKa ID', 'pKa (exp)', 'graph pka 1', 'graph pka 2', 'graph pka 3']
pka_graph = pka_graph[order]
pka_graph
```

```python
pka_graph['Epik Scan'] = list(Epik_Scan['pKa (calc)'])
pka_graph['Epik Microscopic'] = list(Epik_Micro['pKa (calc)'])
pka_graph['ACD/pKa classic'] = list(ACD['pKa (calc)'])
pka_graph['MoKa'] = list(MoKa['pKa (calc)'])
pka_graph['Chemicalize v18.23'] = list(Chemicalize['pKa (calc)'])
```

```python
pka_graph
```

```python
pka_graph.to_csv('./Predict_result/all_SAMPL6_result.csv',index = False)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
print(mean_squared_error(list(pka_graph['pKa (exp)']),list(pka_graph['graph pka 1']))**0.5)
print(mean_absolute_error(list(pka_graph['pKa (exp)']),list(pka_graph['graph pka 1'])))
print(r2_score(list(pka_graph['pKa (exp)']),list(pka_graph['graph pka 1'])))
```
